```python
import pandas as pd 
```

```python
a_df = pd.read_csv("/content/neo_v2.csv")
#импортируем базу данных
```

```python
df = a_df.drop("orbiting_body",  axis = 1)
df = df.drop("name", axis = 1)
df.head()
#Для выполнения численного анализа - оставляем столбцы только с численными типами данных 
```

```python
from sklearn.preprocessing import StandardScaler 
import numpy as np
X = df.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet


#Преобразовываем таблицу данных в массив. Стандартизируем данные массива.

# Стандартизация – это метод масштабирования, при котором данные не масштабируются 
# путем преобразования статистического распределения данных в следующий формат:

# преобразуем данные для их центрирования, удаляя среднее значение каждой функции, 
# а затем масштабируем ее, деля непостоянные характеристики на их стандартное отклонение.

# Применить функцию к набору данных с помощью функции fit_transform().
```

```python
from sklearn.cluster import KMeans
clusterNum = 3 
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(Clus_dataSet)
labels = k_means.labels_
print(labels)

# Применение метода K-means
# n_init умолчанию: 10 :
# Количество раз, когда алгоритм k-средних будет выполняться с разными начальными значениями центроида. 
# Конечными результатами будут наилучшие результаты n_init последовательных запусков с точки зрения инерции.


# Эти KMeans данные алгоритмы кластеров пытаются отдельными образцы в п групп 
# одинаковой дисперсии, сводя к минимуму критерия , известный как инерция или 
# внутри-кластера сумм квадратов . Этот алгоритм требует указания количества 
# кластеров. Он хорошо масштабируется для большого количества образцов и используется в 
# широком диапазоне областей применения во многих различных областях.

# Алгоритм k-средних делит набор образцы в непересекающиеся кластеры,каждый из которых 
# описывается средним образцов в кластере. Средние значения обычно называют «центроидами» 
# кластера; обратите внимание, что это, как правило, не баллы из 
# , хотя они живут в одном пространстве.

# Алгоритм K-средних нацелен на выбор центроидов, которые минимизируют 
# инерцию или критерий суммы квадратов внутри кластера :
```

Коллеряции 


Pearson’s correlation 
Если коэффициент корреляции равен -1, это указывает на сильную отрицательную связь
Если коэффициент корреляции равен 0, это указывает на отсутствие связи.
Если коэффициент корреляции равен 1, это указывает на сильную положительную взаимосвязь.


Корреляция Пирсона может вычислить ТОЛЬКО линейную связь между двумя непрерывными переменными (отношение линейно только тогда, когда изменение одной переменной связано с пропорциональным изменением другой переменной)


Корреляция Спирмена может оценить монотонную связь между двумя переменными - непрерывной или порядковой, и она основана на ранжированных значениях для каждой переменной, а не на необработанных данных.

```python
corr_pearson = df[features_num].corr(method='pearson')
corr_spearman = df[features_num].corr(method='spearman')

plt.figure(figsize=(18,6))
ax1 = plt.subplot(1,2,1)
sns.heatmap(corr_pearson, annot=True, cmap='RdYlGn', vmin=-1, vmax=+1)
plt.title('Pearson Correlation')

ax2 = plt.subplot(1,2,2, sharex=ax1)
sns.heatmap(corr_spearman, annot=True, cmap='RdYlGn', vmin=-1, vmax=+1)
plt.title('Spearman Correlation')
plt.show()
```
